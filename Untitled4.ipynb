```python
import nltk
from nltk.chat.util import Chat, reflections

# Download (safe even if already installed)
nltk.download('punkt')

# Define patterns
pairs = [
    [r"hi|hello|hey", ["Hello!", "Hi there!", "Hey!"]],
    [r"what is your name ?", ["I am an NLP chatbot."]],
    [r"how are you ?", ["I'm doing great!", "All good!"]],
    [r"sorry(.*)", ["No problem!", "It's okay!"]],
    [r"(.*)help(.*)", ["I can help you."]],
    [r"bye", ["Goodbye!", "See you later!"]],
]

'''pairs = [
    [r"hi|hello|hey", ["Hello!", "Hi there!", "Hey!"]],
    [r"what is your name ?", ["I am an NLP chatbot."]],
    [r"how are you ?", ["I'm doing great!", "All good!"]],
    [r"sorry(.*)", ["No problem!", "It's okay!"]],
    [r"my name is (.*)", ["Nice to meet you %1!"]],
    [r"i am (.*)", ["Why are you %1?", "How long have you been %1?"]],
    [r"(.*) help (.*)", ["I can help you."]],
    [r"what is (.*)", ["%1 is something interesting!"]],
    [r"bye", ["Goodbye!", "See you later!"]],
    [r"(.*)", ["I didn't understand that. Can you rephrase?"]],  # keep last
]'''
# Create chatbot
chat = Chat(pairs, reflections)

print("Chatbot ready!")

while True:
    msg = input("You: ")

    if msg.lower() == "bye":
        print("Bot: Goodbye!")
        break

    print("Bot:", chat.respond(msg))
```
